```python
def read_meter(row: pd.Series) -> pd.Series:
    if not pd.isna(row['gasmeasurement']):
        row['gas'] = row['gasmeasurement']['meter']
    if not pd.isna(row['electricitymeasurement']):
        row['elec'] = row['electricitymeasurement']['meter']
    return row
```

```python
# Extract data between 17 and 19h
# create 'time' and 'date' columns
# calculate usage for type = gas/elec
# pivot table

def readings_to_usage(df: pd.DataFrame, type: str) -> pd.DataFrame:
    df = df.copy()
    df['datetime'] = pd.to_datetime(df['datetime'], format='ISO8601', utc=False).dt.tz_localize(None)
    df = df[df['datetime'].dt.time.between(time(15), time(21), inclusive='both')]
    df['keydate'] = df['activationkey'] + '_' + df['datetime'].dt.date.astype(str)
    df['time'] = df['datetime'].dt.time
    df['date'] = df['datetime'].dt.date
    df.sort_values(by=['keydate', 'time'], inplace=True)

    # calculate usage
    
    df = df.apply(read_meter, axis=1)
    df = df[~pd.isna(df[type])]
    df['usage'] = df.groupby('keydate')['elec'].diff()

    df = df.pivot_table(
        index=['activationkey', 'date'],
        columns='time',
        values='usage'
    )
    return df
```

```python
def plot_timeseries_by_activationkey(df, figsize=(15, 8), alpha=0.6):
    """
    Plots all rows of a MultiIndex DataFrame (activationkey, date) as time series,
    using a unique color per activationkey. Values are normalized per row to [0,1].

    Parameters:
    -----------
    df : pandas.DataFrame
        MultiIndex DataFrame with (activationkey, date) as index and datetime.time columns
    figsize : tuple
        Size of the matplotlib figure
    alpha : float
        Transparency of the lines
    """
    # Prepare color map
    unique_keys = df.index.get_level_values('activationkey').unique()
    colors = plt.cm.tab20(np.linspace(0, 1, len(unique_keys)))

    # Convert time objects to minutes since midnight for plotting
    x_values = [(t.hour * 60 + t.minute) for t in df.columns]

    plt.figure(figsize=figsize)

    # Plot each activation key
    for idx, key in enumerate(unique_keys):
        key_data = df.loc[key]
        color = colors[idx]

        # Plot each date for this key
        for date, row in key_data.iterrows():
            # Normalize the row values to [0,1]
            row_min = row.min()
            row_max = row.max()

            # Only normalize if there's a difference between min and max
            # if row_max > row_min:
            #     normalized_values = (row - row_min) / (row_max - row_min)
            # else:
            #     normalized_values = row - row_min  # Will be all zeros if all values are the same
            normalized_values = row
            plt.plot(x_values, normalized_values, color=color, alpha=alpha,
                    label=key if date == key_data.index[0] else "",
                    marker='o', markersize=4)

    # Customize the plot
    plt.xlabel('Time of Day')
    plt.ylabel('Normalized Value (0-1)')
    plt.title('Normalized Time Series by Activation Key')
    plt.grid(True, alpha=0.3, linestyle='--')

    # Format x-axis ticks to show actual times
    time_labels = [t.strftime('%H:%M') for t in df.columns]
    plt.xticks(x_values, time_labels, rotation=45)

    # Set y-axis limits with a small padding
    # plt.ylim(-0.05, 1.05)

    # Add legend
    plt.legend(title='Activation Key', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout()
    plt.show()
```

```python
class Model:

    def __init__(self, gas_users: pd.DataFrame, induction_users: pd.DataFrame) -> None:
        self.gas = gas_users
        self.ind = induction_users

    def train(self):
        features_induction = self.collect_features(self.ind)
        features_gas = self.collect_features(self.gas)

        # Add labels
        features_induction['cooking_method'] = 'induction'
        features_gas['cooking_method'] = 'gas'
        
        # Combine datasets
        features_combined = pd.concat([features_induction, features_gas])

        # Prepare for modeling
        X = features_combined.drop(['cooking_method'], axis=1)
        y = features_combined['cooking_method']
        
        # Scale features
        self.scaler = StandardScaler()
        X_scaled = self.scaler.fit_transform(X)
        
        # Split data
        X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
        
        # Train model
        self.rf_model = RandomForestClassifier(n_estimators=200, random_state=42)
        self.rf_model.fit(X_train, y_train)
        
        # Print performance
        y_pred = self.rf_model.predict(X_test)
        print("\nModel Performance (User-Level Classification):")
        print(classification_report(y_test, y_pred))

    def predict(self, df: pd.DataFrame) -> tuple:
        feature_dict = self._create_features(df)
        user_features = pd.DataFrame([feature_dict])
    
        X_scaled = self.scaler.transform(user_features)
        prediction = self.rf_model.predict(X_scaled)[0]
        probability = self.rf_model.predict_proba(X_scaled)[0]
    
        return prediction, max(probability)
    
    def collect_features(self, df: pd.DataFrame) -> pd.DataFrame:
        features = []
        for activationkey, user_data in df.groupby('activationkey'):
            features.append(self._create_features(user_data))
        return pd.DataFrame(features)
    
    def _create_features(self, df: pd.DataFrame) -> dict:
        time_cols = [col for col in df.columns if ':' in str(col)]
        consumption_data = df[time_cols]
        
        # Time period splits
        early_evening = consumption_data[[col for col in time_cols if col.hour == 15 or col.hour == 16]]
        dinner_time = consumption_data[[col for col in time_cols if col.hour == 17 or col.hour == 18]]
        late_evening = consumption_data[[col for col in time_cols if col.hour == 19 or col.hour == 20]]
        
        # Calculate daily patterns
        daily_peaks = consumption_data.max(axis=1)
        daily_means = consumption_data.mean(axis=1)
        daily_totals = consumption_data.sum(axis=1)
        
        feature_dict = {
            'max_consumption': consumption_data.max().max(),
            'avg_daily_peak': daily_peaks.mean(),
            'std_daily_peak': daily_peaks.std(),
            'mean_consumption': consumption_data.mean().mean(),
            'median_consumption': consumption_data.median().median(),
            'consumption_std': consumption_data.std().std(),
            'avg_daily_total': daily_totals.mean(),
            'std_daily_total': daily_totals.std(),
            'peak_to_mean_ratio': daily_peaks.mean() / daily_means.mean(),
            'active_periods_ratio': (consumption_data > 50).sum().sum() / consumption_data.size,
            'early_evening_avg': early_evening.mean().mean(),
            'dinner_time_avg': dinner_time.mean().mean(),
            'late_evening_avg': late_evening.mean().mean(),
            'zero_periods_ratio': (consumption_data == 0).sum().sum() / consumption_data.size,
            'peak_time_consistency': consumption_data.idxmax(axis=1).mode().count() / len(consumption_data),
        }
        return feature_dict
```

```python
def create_user_features(df):
    time_cols = [col for col in df.columns if ':' in str(col)]
    
    features = []
    for activationkey, user_data in df.groupby('activationkey'):
        consumption_data = user_data[time_cols]
        
        # Time period splits
        early_evening = consumption_data[[col for col in time_cols if col.hour == 15 or col.hour == 16]]
        dinner_time = consumption_data[[col for col in time_cols if col.hour == 17 or col.hour == 18]]
        late_evening = consumption_data[[col for col in time_cols if col.hour == 19 or col.hour == 20]]
        
        # Calculate daily patterns
        daily_peaks = consumption_data.max(axis=1)
        daily_means = consumption_data.mean(axis=1)
        daily_totals = consumption_data.sum(axis=1)
        
        feature_dict = {
            'activationkey': activationkey,
            'max_consumption': consumption_data.max().max(),
            'avg_daily_peak': daily_peaks.mean(),
            'std_daily_peak': daily_peaks.std(),
            'mean_consumption': consumption_data.mean().mean(),
            'median_consumption': consumption_data.median().median(),
            'consumption_std': consumption_data.std().std(),
            'avg_daily_total': daily_totals.mean(),
            'std_daily_total': daily_totals.std(),
            'peak_to_mean_ratio': daily_peaks.mean() / daily_means.mean(),
            'active_periods_ratio': (consumption_data > 50).sum().sum() / consumption_data.size,
            'early_evening_avg': early_evening.mean().mean(),
            'dinner_time_avg': dinner_time.mean().mean(),
            'late_evening_avg': late_evening.mean().mean(),
            'zero_periods_ratio': (consumption_data == 0).sum().sum() / consumption_data.size,
            'peak_time_consistency': consumption_data.idxmax(axis=1).mode().count() / len(consumption_data),
            'num_days': len(user_data)
        }
        features.append(feature_dict)
    
    return pd.DataFrame(features)

# Create features for both types
features_induction = create_user_features(df_induction)
features_gas = create_user_features(df_gas)

# Add labels
features_induction['cooking_method'] = 'induction'
features_gas['cooking_method'] = 'gas'

# Combine datasets
features_combined = pd.concat([features_induction, features_gas])

print("Number of unique users:")
print(f"Induction: {len(features_induction)}")
print(f"Gas: {len(features_gas)}")

# Visualize key differences
plt.figure(figsize=(15, 5))

# Daily consumption patterns
plt.subplot(1, 3, 1)
sns.boxplot(x='cooking_method', y='avg_daily_total', data=features_combined)
plt.title('Average Daily Consumption')
plt.ylabel('Consumption (W)')

# Peak to mean ratio
plt.subplot(1, 3, 2)
sns.boxplot(x='cooking_method', y='peak_to_mean_ratio', data=features_combined)
plt.title('Peak to Mean Ratio')

# Time period comparison
plt.subplot(1, 3, 3)
sns.boxplot(x='cooking_method', y='active_periods_ratio', data=features_combined)
plt.title('Active Periods Ratio')

plt.tight_layout()
plt.show()

# Prepare for modeling
X = features_combined.drop(['activationkey', 'cooking_method'], axis=1)
y = features_combined['cooking_method']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# Train model
rf_model = RandomForestClassifier(n_estimators=200, random_state=42)
rf_model.fit(X_train, y_train)

# Print performance
y_pred = rf_model.predict(X_test)
print("\nModel Performance (User-Level Classification):")
print(classification_report(y_test, y_pred))

# Feature importance
feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='importance', y='feature', data=feature_importance.head(10))
plt.title('Top 10 Most Important Features for User Classification')
plt.xlabel('Importance')
plt.tight_layout()
plt.show()

# Print detailed statistics
print("\nDetailed User Statistics by Cooking Method:")
for method in ['induction', 'gas']:
    subset = features_combined[features_combined['cooking_method'] == method]
    print(f"\n{method.upper()} Users:")
    print(f"Average Daily Consumption: {subset['avg_daily_total'].mean():.2f}W")
    print(f"Average Peak Consumption: {subset['max_consumption'].mean():.2f}W")
    print(f"Active Periods Ratio: {subset['active_periods_ratio'].mean():.2%}")
    print(f"Peak to Mean Ratio: {subset['peak_to_mean_ratio'].mean():.2f}")
    print(f"Time of Day Averages:")
    print(f"  Early Evening: {subset['early_evening_avg'].mean():.2f}W")
    print(f"  Dinner Time: {subset['dinner_time_avg'].mean():.2f}W")
    print(f"  Late Evening: {subset['late_evening_avg'].mean():.2f}W")
```

```python
# Calculate nightly and daily usage (between 23 and 06h)
# create 'time' and 'date' columns
# select readings at 6h and 23h only and calculate difference
# assign 6h differences to 'night' column and 23h diffs to 'day' column
# check if night > day 

def readings_to_nightly_usage(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df['datetime'] = pd.to_datetime(df['datetime'], format='ISO8601', utc=False).dt.tz_localize(None)
    df['keydate'] = df['activationkey'] + '_' + df['datetime'].dt.date.astype(str)
    df['time'] = df['datetime'].dt.time
    df['date'] = df['datetime'].dt.date
    df.sort_values(by=['keydate', 'time'], inplace=True)

    # keep relevant time readings only
    df = df[(df['time'] == time(23)) | (df['time'] == time(6))]
    
    # calculate usage
    
    df = df.apply(read_meter, axis=1)
    df = df[~pd.isna(df['elec'])]

    df = df[['activationkey', 'date', 'time', 'keydate', 'elec']].copy()
    df = df.sort_values(by=['keydate', 'time'])
    
    df['usage'] = df.groupby('activationkey')['elec'].diff()
    night_usage = df[df['time'] == time(6)].set_index('keydate')['usage']
    df['night'] = df['keydate'].map(night_usage)
    day_usage = df[df['time'] == time(23)].set_index('keydate')['usage']
    df['day'] = df['keydate'].map(day_usage)
    df.drop_duplicates(subset='keydate', inplace=True)

    df['nightusage'] = (df['night'] > df['day']) & (df['night'] > 2000)
    df = df[['activationkey', 'date', 'night', 'day', 'nightusage']]
    
    return df
```

```python
def bin_usage(df: pd.DataFrame, name: str, show: bool = False):
    bins = [0, 1, 8, 14, df['sum'].max() + 1]
    colors = ['green', 'yellow', 'orange', 'red']
    labels = ['0', '1-7', '8-13', '14+']
    
    # Get histogram data
    counts, bin_edges = np.histogram(df['sum'], bins=bins)
    total = counts.sum()
    
    fig = plt.figure(figsize=(10, 6))
    
    # Plot each bin as a separate bar
    bars = []
    for i in range(len(counts)):
        bar = plt.bar(
            x=labels[i],
            height=counts[i],
            color=colors[i],
            edgecolor='black'
        )
        bars.append(bar)
        
        # Add percentage on top of each bar
        percentage = (counts[i] / total) * 100
        plt.text(
            x=i,  # x position (bar index)
            y=counts[i] + total * 0.01,  # y position (slightly above bar)
            s=f'{percentage:.1f}%',  # text with 1 decimal place
            ha='center',  # horizontal alignment
            va='bottom',  # vertical alignment
            fontweight='bold'
        )
    
    plt.title(f"Nachtverbruik in juni 2024 - {name}")
    plt.xlabel('Aantal dagen met nachtverbruik')
    plt.ylabel('Aantal huishoudens')
    _ = fig.savefig(f'../data/nachtverbruik_juni_2024_bins_{name}.png', dpi=300, bbox_inches='tight')
```
